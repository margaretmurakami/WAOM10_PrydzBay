```python
xgrid = ds.variables["Xgrid"].values[:,pol_locs]
ygrid = ds.variables["Ygrid"].values[:,pol_locs]
```

```python
xgrid.shape
```

```python
firstx = [xgrid[:,i][np.isfinite(xgrid[:,i])][0] for i in range(len(xgrid[0]))]
firsty = [ygrid[:,i][np.isfinite(ygrid[:,i])][0] for i in range(len(ygrid[0]))]
```

```python
firsth = ds1.h.values
firsth.shape
```

```python
touse = np.array([],dtype=int)
for i in range(len(firstx)):
    myy = int(firsty[i])
    myx = int(firstx[i])
    thisdepth = firsth[myy][myx]
    if thisdepth <= 1000:
        if firstx[i]>478:
            touse = np.append(touse,i)
touse.shape
```

```python
pollocs2 = pol_locs[touse]
pollocs2
```
